````python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt

import numpy as np
import patsy
import warnings
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.interpolate import interp1d
import scipy.stats as ss 
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
````

````python
df = pd.read_csv('./input/previsao_de_renda_II.csv', index_col=0)
df.head()
````

````python
df['tempo_emprego'].fillna(value=df['tempo_emprego'].mean(), inplace=True)
````

````python
df.isna().sum()
````

````python
df_train = df[:'2015-12-01']
df_test = df['2016-01-01':]
````

````python
# Copie os DataFrames, se necessrio, para evitar a modificao do DataFrame original
df_train_copy = df_train.copy()
df_test_copy = df_test.copy()

# Crie uma nova coluna 'log_renda' nos DataFrames copiados
df_train_copy['log_renda'] = np.log(df_train_copy['renda'])
df_test_copy['log_renda'] = np.log(df_test_copy['renda'])

# Se estiver tudo correto, voc pode substituir os DataFrames originais pelos DataFrames copiados
df_train = df_train_copy
df_test = df_test_copy

# Ignore os warnings durante o processo de atribuio
warnings.filterwarnings('ignore')
````

````python
X1 = patsy.dmatrices('''log_renda ~ C(posse_de_imovel)
                    + qtd_filhos 
                    + idade
                    + tempo_emprego
                    + 1''', df_train)

res = smf.ols(X1, data=df_train).fit_regularized(method = 'elastic_net' 
                         , refit = True
                         , L1_wt = 1
                         , alpha = 0.05)
````

````python
res.summary()
````

````python
X1 = patsy.dmatrices('''log_renda ~ C(posse_de_imovel)
                    + tempo_emprego
                    + 1''', df_train)

res = smf.ols(X1, data=df_train).fit_regularized(method = 'elastic_net' 
                         , refit = True
                         , L1_wt = 1
                         , alpha = 0.05)
````

````python
res.summary()
````

## Multicolinearidade

Avalie se h questes relacionadas a multicolinearidade atravs de pelo menos:

- Matriz de correlao de Spearman
- VIF

````python
df_train_cut = df_train[['posse_de_imovel', 'tempo_emprego', 'log_renda']]

df_train_cut.head(2)
````

````python
# Matriz de correlao de Spearman

df_train_cut.corr(method='spearman')
````

````python
# VIF

variaveis = ['tempo_emprego', 'posse_de_imovel']

dum = pd.get_dummies(df_train_cut[variaveis], drop_first=True)
dum.head(2)
````

````python
vif(dum.values, 1)
````

````python
vars_vif = pd.DataFrame()
vars_vif["VIF Factor"] = [vif(dum, i) for i in range(dum.shape[1])]
vars_vif["Feature"] = dum.columns

vars_vif.round(2)
````
